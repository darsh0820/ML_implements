#Iris dataset

#Load Dataset

```python
from sklearn import datasets
import pandas as pd
iris = datasets.load_iris()
```

#Dataset Features

##Input Features

```python
print(iris.feature_names)
```

##Output Features

```python
print(iris.target_names)
```

# Data Preparation

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df.head(10)
```

##Data separation as x and y

```python
Y = df['target']
Y.head(10)
```

```python
X = df.drop('target', axis=1)
X.head(10)
```

##Data Splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train.head(10)
```

```python
X_test.head(10)
```

#Model Building

##Random Forest Classification

###Training the model

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, Y_train)
```

###Applying the model to make prediction


```python
y_rfc_train_pred = rfc.predict(X_train)
y_rfc_test_pred = rfc.predict(X_test)
```

```python
y_rfc_train_pred
```

```python
y_rfc_test_pred
```

```python
rfc.predict_proba(X_test)
```

###Model Performance

```python
rfc_score_test = rfc.score(X_test, Y_test)
rfc_score_train = rfc.score(X_train, Y_train)
```

```python
print('RF (Test)',rfc_score_test)
print('RF (Train)',rfc_score_train)
```

```python
rfc_results = pd.DataFrame([['Random Forest',rfc_score_train,rfc_score_test]],
               columns = ['Model', 'Training Accuracy','Testing Accuracy'])
rfc_results
```

#Model Comparison

```python
df_models = pd.concat([rfc_results])
df_models.reset_index(drop = True)
```

#Data Visualization of prediction results


Importing Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

##Confusion Matrix

###Random Forest Classification

```python
#Random Forest Classification
rfc_cm = confusion_matrix(Y_test, y_rfc_test_pred)
rfc_cm_df = pd.DataFrame(rfc_cm, index = ['setosa','versicolor','virginica'], columns = ['setosa','versicolor','virginica'])
plt.figure(figsize=(6,4))
sns.heatmap(rfc_cm_df, annot=True)
plt.ylabel('Actual Values')
plt.xlabel('Predicted Values')
plt.suptitle("Random Forest Classification")
plt.title("Accuracy: " + str(round(rfc_score_test,3)))
plt.show()
```

##Graph

```python
rfc_df_pred = pd.DataFrame(X_test)
rfc_df_pred['target'] = y_rfc_test_pred
rfc_df_pred['flower_name'] = rfc_df_pred['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

plt.figure(figsize=(5.5,4))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='flower_name', data=rfc_df_pred)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Random Forest Classification - Predicted Results')
plt.show()
```
